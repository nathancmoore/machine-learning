```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
boston = datasets.load_boston()
```

```python
med_home_val = np.median(boston['target'])
avg_home_val = np.mean(boston['target'])
print('Median Home Value: ', med_home_val)
print('Mean Home Value: ', avg_home_val)
```

* *Median/Mean Home Value*
    * The median home value in suburban Boston approximately 1,000 USD less than the mean. This suggests that the distribution of home values in suburban Boston are skewed right — with a few high-priced home values pulling the average up, while maintaining the median's lower value. 

```python
crime_riv = []
crime_no_riv = []
pupil_capita_riv = []
pupil_capita_no_riv = []
for case in boston.data:
    if case[3] == 1:
        crime_riv.append(case[0])
        pupil_capita_riv.append(case[10])
    else:
        crime_no_riv.append(case[0])
        pupil_capita_no_riv.append(case[10])
plt.scatter(pupil_capita_no_riv, crime_no_riv, s=15, c="red")
plt.scatter(pupil_capita_riv, crime_riv, s=15, c="blue")
plt.ylabel("Crime per Capita")
plt.xlabel("Pupil-to-Teacher Ratio")
plt.title("Crime/Pupil-To-Teacher Ratio in the Boston Suburbs")
plt.show()
```

* *Crime & Pupil-to-Student ratio*
    * The crime rate is higher in schools where the pupil-to-teacher ratio is higher, with more students per teacher. Additionally, there are fewer schools total that are near the river, but those that are riverside properties have lower than average crime rates, and have a lower pupil-to-teacher ratio. This suggests that riverside schools are of higher value, and have more funding — and therefore lower crime rates. 

```python
black_czns = []
dst_to_center = []
for case in boston.data:
    black_czns.append(case[11])
    dst_to_center.append(case[7])
plt.scatter(black_czns, dst_to_center, s=10, c="green")
plt.xlabel("Proportion of Black Citizens/Town")
plt.ylabel("Distance to Employment Center")
plt.title("Proportion of Black Citizens vs Dist to Employment Centers")
plt.show()
```

* *Black Citizens/Town & Employment Centers*
    * The data above shows that there is a correlation between distance to employment centers and % of town population made up by black citizens. If one is to extrapolate this (presuming causation) this suggests that towns further away from employment centers are more likely to have a higher black population.
